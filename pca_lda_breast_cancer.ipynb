# Scaling the data

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

X = pd.DataFrame(scaler.fit_transform(X), columns = X.columns)
```

```python
X
```

# Creating Covariance Matrix and finding Eigenvalues and Eigen Vectors


```python
X.iloc[:, 1].values
```

```python
cov_mat = np.cov(X.T)
```

```python
cov_mat
```

```python
cov_mat.shape
```

```python
eigenvalues, eigenvectors = np.linalg.eig(cov_mat)
```

```python
print(f"Eigenvalues : \n {eigenvalues},\n Eigenvectors: \n{eigenvectors}")
```





* The explained variance is the proportion of variance in the data that is accounted for by each principal component.
* We can sort the eigenvalues in descending order and calculate the proportion of the total variance that is explained by each principal component. The explained variance of the i-th principal component is given by:

```python
sort_eigen_values = sorted(eigenvalues)
```

```python
eigenvalues
```

```python
# Calculate total variance
total_variance = np.sum(eigenvalues)

# Calculate explained variances
explained_variances = eigenvalues / total_variance
explained_variances
```

```python
len(eigenvalues / (np.sum(eigenvalues)) * 100)
```

```python
explained_variance = []
c = 0
for i in range(len(eigenvalues)):
    c = c + np.round((eigenvalues[i] / (np.sum(eigenvalues)) * 100), 3)
    while c < 92:
        explained_variance.append(c)
        print(f"At {i} PCA, Explained Variance is {c}")
        break
                        
```

```python
from sklearn.decomposition import PCA
```

```python
principal = PCA(n_components=6)
principal.fit(X)
principal = principal.transform(X)
principal
```

```python
sns.lineplot(x=explained_variance,  y=range(1,len(explained_variance)+1))
plt.title('Explained Variance Plot', fontsize=18)
plt.xlabel('Explained Variance', fontsize=16)
plt.ylabel('Principal Components',fontsize=16)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.75, random_state=0 )
```

```python
y.value_counts()
```

```python
# Algorithms
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.pipeline import Pipeline

#Performance Model

from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.metrics import classification_report
```

```python
algorithms = [
    'Logistic Regression', LogisticRegression(),
    "KNN", KNeighborsClassifier(n_neighbors=5),
    "Linear SVM", SVC(kernel="linear", C=0.030),
    "Gradient Boosting", GradientBoostingClassifier(learning_rate=0.10, n_estimators=100),
    "Decision Tree", DecisionTreeClassifier(max_depth=7, criterion= "entropy", random_state=0),
    "Random_Forest", RandomForestClassifier(max_depth=5, n_estimators=100, random_state=0)
]

for i in range(0, len(algorithms), 2):
    name = algorithms[i]
    algo = algorithms[i + 1]
    pipe = Pipeline([("pca", PCA(n_components=6)), (name, algo)])
    pipe.fit(X_train, y_train)
    y_pred = pipe.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    
    train_yhat = pipe.predict(X_train)
    train_acc = accuracy_score(y_train, train_yhat)
    print(f"Accuracy Score for training {name} : {train_acc * 100}")
    

    print(f"Accuracy Score for  {name} : {accuracy * 100}")
```
